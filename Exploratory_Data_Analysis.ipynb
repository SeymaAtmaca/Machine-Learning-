```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.style.use('ggplot')
iris=datasets.load_iris()

#iris.target_names

x=iris.data
y=iris.target


veri=pd.DataFrame(x,columns=iris.feature_names)
```

```python
pd.plotting.scatter_matrix(veri,c=y,figsize=[8,8],s=50)
plt.show()
```

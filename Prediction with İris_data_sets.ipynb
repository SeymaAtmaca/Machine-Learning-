```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
plt.style.use('ggplot')
iris=datasets.load_iris()


x=iris.data
y=iris.target

data=pd.DataFrame(x,columns=iris.feature_names)

pd.plotting.scatter_matrix(data,c=y,figsize=[10,10],s=50)
plt.show()
```

```python
petal_length=data['petal length (cm)']
petal_width=data['petal width (cm)']

#hue olusturma
z=iris.target

data['target_names']=pd.DataFrame(z)
data['target_names']=data['target_names'].replace({0 : 'setosa',1 : 'versicolor',2 : 'virginica'})

sns.scatterplot(petal_length,petal_width,hue=data['target_names'])
plt.show()
```
